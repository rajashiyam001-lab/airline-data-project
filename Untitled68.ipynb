#### 1. Coût moyen du carburant par avion de la flotte (en millions USD)


```python
df["fuel_cost_per_aircraft_m"] = df["fuel_cost_usd_m"] / df["fleet_size"]
```

#### 2. Ratio efficacité couverture (Savings / Fuel Cost)

```python
df["hedging_efficiency_ratio"] = np.where(df["fuel_cost_usd_m"] > 0, 
                                          df["hedge_savings_usd_m"] / df["fuel_cost_usd_m"], 0)
```

```python
print("\n--- KPI par phase de conflit ---")

kpi_segments = (
    df.groupby("conflict_phase")
      .agg({
          "jet_fuel_usd_barrel": "mean",
          "fuel_cost_pct_revenue": "mean",
          "profit_margin_pct": "mean",
          "fleet_size": "median"
      })
      .rename(columns={
          "jet_fuel_usd_barrel": "prix_jetfuel",
          "fuel_cost_pct_revenue": "cout_carburant_pct",
          "profit_margin_pct": "marge_pct",
          "fleet_size": "taille_flotte"
      })
      .reset_index()
)

print(kpi_segments)
```

### Modélisation Prédictive (Régression Linéaire)

```python
X = df[["fuel_cost_pct_revenue"]]
y = df["profit_margin_pct"]

# Création du modèle
model = LinearRegression()

# Entraînement
model.fit(X, y)

# Prédictions
y_pred = model.predict(X)

# Résultats
print("\n--- Régression linéaire ---")

coef = model.coef_[0]
intercept = model.intercept_
r2 = r2_score(y, y_pred)

print("Coefficient :", round(coef, 4))
print("Intercept :", round(intercept, 2))
print("R² :", round(r2, 4))
```

### Data Visualization 

```python
sns.set_theme(style="whitegrid")

fig, ax = plt.subplots(figsize=(10, 6))

# Nuage de points
sns.scatterplot(
    data=df,
    x="fuel_cost_pct_revenue",
    y="profit_margin_pct",
    hue="conflict_phase",
    ax=ax
)

# Ligne de régression
x_range = np.linspace(
    df["fuel_cost_pct_revenue"].min(),
    df["fuel_cost_pct_revenue"].max(),
    100
).reshape(-1, 1)

y_pred_line = model.predict(
    pd.DataFrame(x_range, columns=["fuel_cost_pct_revenue"])
)

ax.plot(x_range, y_pred_line, color="black")

# Ligne seuil
ax.axhline(0, linestyle="--")

# Titres
ax.set_title("Impact du coût carburant sur la marge")
ax.set_xlabel("Coût carburant (%)")
ax.set_ylabel("Marge profit (%)")

plt.tight_layout()
plt.show()
```
